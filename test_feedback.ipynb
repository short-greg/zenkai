```python
import torch.nn as nn
import torch
import zenkai
import sklearn

import torch.utils.data as torch_data
```

```python
import sklearn.linear_model


torch.manual_seed(1)

class Net(nn.Module):

    def __init__(self, in_features, mid_features, out_features):
        super().__init__()

        self.layer1 = nn.Linear(in_features, mid_features)
        self.layer2 = zenkai.feedback.FALearner(
            nn.Linear(mid_features, out_features), 
            nn.Linear(mid_features, out_features),
            zenkai.optimf('Adam', lr=1e-3)
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:

        y = self.layer1(x)
        y = torch.sigmoid(y)
        return self.layer2(y)


x = torch.randn(1000, 4)
base = Net(4, 8, 6)
t = base(x).detach()
trainable = Net(4, 8, 6)
zenkai.set_lmode(
    trainable, zenkai.LMode.WithStep
)
dataset = torch_data.TensorDataset(x, t)

for i in range(100):
    for x_i, t_i in torch_data.DataLoader(dataset, shuffle=True, batch_size=128):

        y_i = trainable(x_i)
        loss = (y_i - t_i).pow(2).mean()
        loss.backward()
        print(loss.item())
```
